```python
def split_gender(gen):
    if type(gen) is float:
        if numpy.isnan(gen):
            return [gen]
    if type(gen) is str:
        gens = gen.split('|')
        return gens[:-1]
```

```python
def split_langs(lang):
    if type(lang) is float:
        if numpy.isnan(lang):
            return [lang]
    if type(lang) is str:
        langs = lang.split('|')
        return langs[:-1]
```

```python
from collections import defaultdict
import time

def int_dict_factory():
    return defaultdict(int)

year_genders = defaultdict(int_dict_factory)
year_langs = defaultdict(int_dict_factory)


#get years
for index, row in df.iterrows():
    pot_dob = row['dob']
    if math.isnan(pot_dob) or pot_dob > 2014:
        continue # we are not including nans
    else:
        dob = int(pot_dob)
        genders = row['genders']
        langs = row['langs']
        for year_dict, vrs in zip([year_genders, year_langs], [genders, langs]):
            for var in vrs:
                try:
                    if math.isnan(var):
                        continue #there isn't a good var here
                except TypeError:
                    pass
                year_dict[dob][var]+=1
                
#abstracted: we want year-gender, but also
#gender-ethnicity -citizenship -place of birth, site-links

for index, row in df.iterrows():
    dob = row['dob']
    dod = row['dod']
    genders = row['genders']
    ethnic_group = row['ethnic_group'] 
    citizenship = row['citizenship']
    place_of_birth = row['place_of_birth'] 
    site_links = row['site_links']
```

```python
import pywikibot

#Tranforming QIDs into English labels.
enwp = pywikibot.Site('en','wikipedia')
wikidata = enwp.data_repository()

def english_label(qid):
    page = pywikibot.ItemPage(wikidata, qid)
    data = page.get()
    return data['labels']['en']

sex_qs = [str(q) for q in yg.columns]
sex_labels = [english_label(sex_q) for sex_q in sex_qs]
yg.columns = sex_labels
```

```python
category_wise_data.loc[:, 'male/female ratio'] = category_wise_data.male/category_wise_data.female
category_wise_data.loc[:, 'female/male ratio'] = 1/category_wise_data['male/female ratio']
```

```python
cat_sorted = category_wise_data.sort('total', ascending=False)
cat_sorted.head()
```

```python
mf = category_wise_data[category_wise_data['male/female ratio'] < np.inf]
# mf.head()
mf = mf[mf['total'] > 19000] # total greater than 1% of total population
mf.sort('male/female ratio', ascending=False).iloc[:, [ 12, 13, 15, 16]].head(15)
```

```python
fm = category_wise_data[category_wise_data['female/male ratio'] < np.inf]
# fm.head()
fm = fm[fm['total'] > 2000] # .1 % of the total population
fm.sort('female/male ratio', ascending=False).iloc[:, [0, 12, 13, 15, 17]].head(15)

## The total population cutoff has to be reduced to show categories with higher ration
## of female to male ratio and also to account for less fraction of women in total population.

##
## NOTE the presence of a category called 'female creature' as occupation. Interestingly, we
## did not find any such classification for males. This could potentially highlight the fact
## that for women, just being women is also considered an occupation. One can also note the
## emergence of categories like model, media professional, translators etc.,
## 
```

```python
# Pearsons' correlation coefficient between number of male and female people
# working in top occupations.

x, y = list(range(1, 100, 10)), [np.corrcoef(cat_sorted['male'][:x], cat_sorted['female'][:x])[0,1] for x in range(1, 100, 10)]
plt.xlabel('Number of top categories')
plt.ylabel('Correlation coefficient')
plt.ylim([.8, 1])
plt.plot(x, y)

## A very high degree of correlation in occupation numbers of male and 
## female across varying number of top categories (by number).


##
## One can say that top categories have equal opportunity chances for
## males and females. However, there are certain areas which are *niche*
## for respective genders. For example, religious servant and warrior
## males against model, media professional for females.
## It can be interesting, however, to see that if females are more likely
## to be present in *soft* occupation as compared to males.
##
```
